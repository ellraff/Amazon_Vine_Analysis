```python
import pandas as pd
```

```python
# Read in the dataframe
f_name = 'Resources/vine_table.csv'
vine_df = pd.read_csv(f_name)
vine_df
```

```python
total_vine_df = vine_df[vine_df['total_votes'] >= 20]
total_vine_df
```

```python
help_total_df = total_vine_df[total_vine_df['helpful_votes']/total_vine_df['total_votes'] >= 0.5]
help_total_df
```

```python
vine_paid_df = help_total_df[help_total_df['vine'] == 'Y']
vine_paid_df
```

```python
vine_unpaid_df = help_total_df[help_total_df['vine'] == 'N']
vine_unpaid_df
```

```python
# Total paid reviews
total_paid = len(vine_paid_df)
total_paid
```

```python
# Total unpaid reviews
total_unpaid = len(vine_unpaid_df)
total_unpaid
```

```python
# Number of 5 star reviews for paid reviews
top_paid = len(vine_paid_df[vine_paid_df['star_rating'] == 5])
top_paid
```

```python
# Number of 5 star reviews for unpaid reviews
top_unpaid = len(vine_unpaid_df[vine_unpaid_df['star_rating'] == 5])
top_unpaid
```

```python
# Percentage of 5 star reviews for unpaid
perc_paid = (top_paid/total_paid) * 100
```

```python
# Percentage of 5 star reviews for unpaid
perc_unpaid = (top_unpaid/total_unpaid) * 100
perc_unpaid
```

```python
reviews = [['Vine Reviews',total_paid, top_paid, perc_paid],
           ['Non-Vine Reviews', total_unpaid, top_unpaid, perc_unpaid]]

reviews_df = pd.DataFrame(reviews, columns = ['Review Source', 'Total Count', '5 Star Rating Count', '% of 5 Star Rating'])
reviews_df
```
